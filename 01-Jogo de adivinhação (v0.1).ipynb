```python
#Backdoor
import random

numero = random.randint(0,10)
tentativas = 1

#Entrada do usuário
user = int(input("Adivinhe o número que estou pensando de 0 a 10!\n"))
while user != numero:
 tentativas += 1

#Processamento
 if user < numero:
  print("Você errou! O número é maior que esse.")
  user = int(input(f"Tente novamente:\n"))

 elif numero < user:
  print("O número é menor que esse.")
  user = int(input(f"Tente novamente:\n"))

 else: user == numero
print(f"Você acertou! O número era {user}.")
print(f"Suas tentativas foram {tentativas}.")
```
