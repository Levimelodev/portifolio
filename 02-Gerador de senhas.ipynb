```python
# Background
import random

letras = list("abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ")
numeros = list("1234567890")
especial = list("!@#$%&*()^?-=+./")
caracteres = letras + numeros + especial
cont = 0
senha = ""

# Entrada do usuário
num = int(input("Quantos caracteres você quer na sua senha?\n"))

# Repetição
while num > cont:
  ger = random.choice(caracteres)
  senha = (senha)+(ger)
  cont = cont+1

print("Senha gerada com êxito.")
print(f"{senha}")
```
